```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats.stats import pearsonr
```

```python
#Import Data
data=pd.read_csv('housing.csv')
```

```python
print(data.info())
print('----------------------------------------------------------------------')
print(data.columns)
```

```python
data.nunique()
```

```python
data.shape
```

```python
op_counts = data['ocean_proximity'].value_counts()
explode=[0.01,0.01,0.01,0.01,0.01]
plt.pie(op_counts, labels=op_counts.index, autopct='%1.1f%%', explode=explode,startangle=90)
plt.title('House Ocean Proximity Distribution')
plt.show()
```

```python
sns.countplot(x='ocean_proximity', data=data, color = 'b')
```

```python
data.plot.scatter(x='median_income', y='median_house_value', title='Scatter plot for Location and Income');
```

```python
# Heatmap for correlation
plt.figure(figsize=(10, 6))
sns.heatmap(data[['housing_median_age', 'population', 'households','total_rooms','total_bedrooms','median_income','median_house_value' ]].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
print(pearsonr(data['median_income'], data['median_house_value']))
```

```python
data['median_house_value'].hist()
```

```python
data.groupby('ocean_proximity')['median_house_value'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Total Values by Location')
plt.ylabel('Average Value')
plt.xlabel('Location')
plt.show();
```

```python
data.groupby('ocean_proximity')['median_income'].min().sort_values(ascending=False).head(10).plot(kind='bar', color ='g')
plt.title('Minimum median income by Location')
plt.ylabel('Average Value')
plt.xlabel('Location')
plt.show();
```

```python
sns.boxplot(x='ocean_proximity', y='housing_median_age', data=data)
plt.title('Prices in various Location');
```

```python
sns.violinplot(x='ocean_proximity', y='housing_median_age', data=data);
```

```python
sns.boxplot(x='ocean_proximity', y='median_house_value', data=data)
plt.title('Prices in various Location');
```
