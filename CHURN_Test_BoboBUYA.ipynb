```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import Data
data=pd.read_csv('churn.csv')
```

```python
data.info()
```

```python
Churn_counts = data['Churn'].value_counts()
explode=[0.01,0.01]
plt.pie(Churn_counts, labels=Churn_counts.index, autopct='%1.1f%%', explode=explode,startangle=90)
plt.title('Churn Distribution')
plt.show()
```

-----------------------------------------------------------------------------------------------------------------

CHURN DISTRIBUTION:
30.4 % of customers are churn.
69.6 % of customers are not churn.

-----------------------------------------------------------------------------------------------------------------

```python
sns.countplot(x='PaymentMethod',hue='Churn', data=data)
plt.xticks(rotation=90);
```

```python
Churn_counts = data['PaymentMethod'].value_counts()
explode=[0.01,0.01,0.01,0.01]
plt.pie(Churn_counts, labels=Churn_counts.index, autopct='%1.1f%%', explode=explode,startangle=90)
plt.title('Payment Method Distribution')
plt.show()
```

```python
#filtering Diabetic Only
cdata = data[data['Churn']=='Yes']

Churn_counts = cdata['PaymentMethod'].value_counts()
explode=[0.01,0.01,0.01,0.01]
plt.pie(Churn_counts, labels=Churn_counts.index, autopct='%1.1f%%', explode=explode,startangle=90)
plt.title('CHURN Payment Method Distribution')
plt.show()
```

----------------------------------------------------------------------------------------------------------------

Even though the payment method distribution seems the same, half of churn [51.3%] uses Electronic Check
